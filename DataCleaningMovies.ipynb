```python
df = pd.merge(df, df1, on = 'movieId', how='inner')
```

```python
df = df.drop(['movieId', 'userId', 'timestamp'], axis = 1)
```

```python
# Group by the data set Based on the Movie title
df = df.pivot_table(df, index=['title', 'genres'], aggfunc = 'mean')
df.reset_index(level=['title', 'genres'], inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
# Check the distribution ratings across the movie

plt.rcParams['figure.figsize'] = (8,3)
plt.style.use('fivethirtyeight')
sns.distplot(df['rating'], color = 'black')
plt.title('Distribution of ratings')
plt.show()
# we can observe from the output that maximum movies has the ratings three to four.
```

```python
# Extract the year from the title of the movie

df['y'] = df['title'].str.split(' ')
df['year'] = df['y'].apply(lambda x: x[-1])
```

```python
df.head()
```

```python
# Lets drop the unnecessary 'y' column

df = df.drop(['y'], axis=1)
df.head()
```

```python
# Lets remove the fisrt brackets from the column y

df['year'] = df['year'].str.strip(')')
df['year'] = df['year'].str.strip('(')
```

```python
df.head()
```

```python
pd.set_option('max_rows', 110)
df['year'].value_counts()
```

```python
# There are some dirty values , lets clean them

df['year'] = df['year'].replace(('2007-', 'Despot', 'Things', '1975-1979', 'Road', ''),('2007', '2016', '2016', '1979', '2002', '1996'))
```

```python
# Convert the year column into integer

df['year'] = df['year'].astype(int)
df.head()
```

```python
df['title'] = df['title'].str.split(' ')
df['title'] = df['title'].apply(lambda x: ' '.join(x[:-1]))
```
